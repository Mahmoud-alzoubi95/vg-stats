```python
import pandas as pd

df_weather = pd.read_csv('./weather.csv')
df_weather.head(5)
```

```python

df_station = pd.read_csv('station.csv')
df_station.head(5)

```

```python
df_trip = pd.read_csv('./trip.csv', error_bad_lines=False)
df_trip
```

```python
# the average trip duration for a borrowed bicycle?
avg = df_trip[['tripduration']].mean()
avg
```

```python
#the most common age of a bicycle-sharer
most_common = (2021 - df_trip['birthyear'].mode())
most_common
```

```python
# Q5 #Given all the weather data here, find the average precipitation per month, and the median precipitation.

Precipitation_In = df_weather['Precipitation_In'].mean()
Precipitation_In
```

```python
# Q5 #Given all the weather data here, find the average precipitation per month, and the median precipitation.

Precipitation_In_meadian = df_weather['Precipitation_In'].median()
Precipitation_In_meadian
```

```python
#What’s the average number of bikes at a given bike station
average_number_of_bikes = df_station['install_dockcount'].mean()
average_number_of_bikes
```

```python
#When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?
# because the deffirence negative
modified = df_station[['modification_date','install_dockcount','current_dockcount']].dropna() #remove any row with no value
modified
modified_current_dockcount = modified['current_dockcount'].sum()
modified_install_dockcount = modified['install_dockcount'].sum()
current_value = modified_current_dockcount - modified_install_dockcount
current_value
```
